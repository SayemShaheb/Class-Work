```python
# Define matrices A and B
import numpy as np
A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

# 1
sum_AB = A + B
print("Element-wise Sum (A + B):\n", sum_AB)

# 2
product_AB = A * B
print("\nElement-wise Product (A * B):\n", product_AB)

# 3
dot_product = np.dot(A, B.T)
print("\nDot Product (A . B^T):\n", dot_product)

#4
determinant_A = np.linalg.det(A)
print("\nDeterminant of A:", determinant_A)
```
