```python
Fruits = ['Apple', 'Banana', 'Cherry']
Last_fruits = Fruits[-1]         # (To find out Last index)
print(Last_fruits)                 #  => Cherry
Fruits.append('Orange')            # (To insert at the last)
print(Fruits)                      #  => ['Apple', 'Banana', 'Cherry', 'Orange']
Fruits.insert(2, 'Mango')          # (To insert at the specific position)
print(Fruits)                      #  => ['Apple', 'Banana', 'Mango', 'Cherry', 'Orange']
Fruits.insert(len(Fruits), 'Berry')# (To insert at the last)
print(Fruits)                      #  => ['Apple', 'Banana', 'Mango', 'Cherry', 'Orange', 'Berry']
Fruits.insert(-1, 'Pineapple')     # (To insert at the Second-to-last position)
print(Fruits)                      #  => ['Apple', 'Banana', 'Mango', 'Cherry', 'Orange', 'Pineapple', 'Berry']
del Fruits[1]                      # (To remove specific index from the list)
print(Fruits)                      #  => ['Apple', 'Mango', 'Cherry', 'Orange', 'Pineapple', 'Berry']
Fruits.remove('Mango')             # (To remove specific item from the list)
print(Fruits)                      #  => ['Apple', 'Cherry', 'Orange', 'Pineapple', 'Berry']
Fruits.pop()                       # (To remove from the last position)
print(Fruits)                      #  => ['Apple', 'Cherry', 'Orange', 'Pineapple']
Count = Fruits.count('Coconut')      # (To count how many coconuts have into the list)
print(Count)                       #  => 0
New_fruits = Fruits.copy()         # (To copy the list)
print(New_fruits)                  #  => ['Apple', 'Cherry', 'Orange', 'Pineapple']
Fruits.clear()
print(Fruits)                      #  => [] (To Clear the list)
                                   # print(Fruits[start : End+1])     =>List Slicing
```

```python
Num = [0,3,9,1,4,6,0,5,4,6,9,1]
ans = 0
for i in range(len(Num)):
  count = 0
  for j in range(i+1, len(Num)):
    if Num[i] == Num[j]:
      count += 1
  ans += count
print(ans)          # => 5
```

```python
Num = [9,0,3,1,4,6,0,5,4,6,9,1,8] #[0,0,1,1,3,4,4,5,6,6,8,9,9]
dic = {}                            # [0: 0]
result = []                         # [1: 2]
sort_num = sorted(Num)              # [3: 4]
for i in range(len(sort_num)):      # [4: 5]
  if sort_num[i] not in dic: #------> [5: 7]  [number: Index]
    dic[sort_num[i]] = i            # [6: 8]
for k in Num:                       # [8: 10]
  if dic[k] not in result:          # [9: 11]
   result.append(dic[k])
print(dic) # =>{0: 0, 1: 2, 3: 4, 4: 5, 5: 7, 6: 8, 8: 10, 9: 11}
print(result)  # =>[11, 0, 4, 2, 5, 8, 7, 10]
```

```python
Num = [9, 0, 3, 1, 4, 6, 5, 8, 2, 7]
target = 10
result = []
dic = {}  # Dictionary to store numbers

for i in Num:
    complement = target - i
    if complement in dic:  # Check if complement exists in the dictionary
        result.append((i, complement))
        print(i, complement)
    dic[i] = True  # Add the current number to the dictionary

# Print result pairs
print(result)
```

```python
# 03
even = 0
odd = 0

print("Enter numbers one by one. Enter 0 to stop.")

while True:
    num = int(input("Enter a number: "))
    if num == 0:
        break
    elif num % 2 == 0:
        even += 1
    else:
        odd += 1

print(f"Total even numbers: {even}")
print(f"Total odd numbers: {odd}")
```

```python
# 07
num = int(input("Enter a number to check if it is prime: "))

if num <= 1:
    print("Not Prime")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime")
    else:
        print("Not Prime")
```

```python
balance = 0
while True:
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = input("Please choose an option : ")
    if choice == "1":
        print(f"Your current balance is: bdt{balance}")

    elif choice == "2":
        try:
            deposit = float(input("Enter amount to deposit: bdt"))
            if deposit <= 0:
                print("Deposit amount must be positive!")
            else:
                balance += deposit
                print(f"Deposited bdt{deposit}. Your new balance is: bdt{balance}")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

    elif choice == "3":
        try:
            withdrawal = float(input("Enter amount to withdraw: bdt"))
            if withdrawal <= 0:
                print("Withdrawal amount must be positive!")
            elif withdrawal > balance:
                print("Insufficient balance! Withdrawal amount exceeds available balance.")
            else:
                balance -= withdrawal
                print(f"Withdrew bdt{withdrawal}. Your new balance is: bdt{balance}")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

    elif choice == "4":
        print("Exiting the banking system")
        break

    else:
        print("Invalid choice! Please select a valid option.")
```
