```python
string = input("Enter a string: ")
reverse =  string[::-1]
print(reverse)

if string == reverse:
    print("palindrom")
else:
    print("not palindrom")
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
num = int(input('Enter a number: '))
print(f"Factorial: {factorial(num)}")
```

```python
start, end = map(int, (input'Enter two integers: ').split())
even_sum = sum(i for i in range(start, end + 1) if i % 2 == 0)
print(f"Sum of even numbers: {even_num}")
```

```python
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, 0):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence[:n]
terms = int(input('Enter the number of terms: '))
print(f"Fibonacci sequence: {fibonacci(terms)}")
```

```python
nums = list(map(int, input('Enter a list of integer: ').split()))
even = [x for x in nums if x % 2 == 0]
odd = [x for x in nums if x % 2 != 0]
print(f"Even numbers: {even}, Odd numbers: {odd}")
```

```python
def remove_duplicates(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]
nums = list(map(int, input('Enter a list of integer: ').split()))
print(f"List without duplicate: {remove_duplicates(nums)}")
```

```python
words = input('Enter a list of word: ').split()
max_word = max(words, key=len)
print(f"Word with maximum lenght: {max_word}")
```
