```python
import pandas as pd
file_path = "/mnt/data/employees.csv"
df = pd.read_csv(file_path)
# Display first few rows
print(df.head())

# 2
average_salary_by_department = df.groupby("DEPARTMENT_ID")["SALARY"].mean()
print(average_salary_by_department)

# 3
highest_paid_department = average_salary_by_department.idxmax()
highest_avg_salary = average_salary_by_department.max()
print(f"Department {highest_paid_department} has the highest average salary: ${highest_avg_salary:.2f}")
```

```python
# Define matrices A and B
import numpy as np
A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

# 1
sum_AB = A + B
print("Element-wise Sum (A + B):\n", sum_AB)

# 2
product_AB = A * B
print("\nElement-wise Product (A * B):\n", product_AB)

# 3
dot_product = np.dot(A, B.T)
print("\nDot Product (A . B^T):\n", dot_product)

# 4
determinant_A = np.linalg.det(A)
print("\nDeterminant of A:", determinant_A)

```

```python
import numpy as np
data = np.random.rand(10000)

# 1
mean_value = np.mean(data)
median_value = np.median(data)
std_dev = np.std(data)
print(f"Mean: {mean_value}, Median: {median_value}, Standard Deviation: {std_dev}")

# 2
min_value = np.min(data)
max_value = np.max(data)
print(f"Min: {min_value}, Max: {max_value}")

# 3
count_greater_than_mean = np.sum(data > mean_value)
print(f"Values greater than mean: {count_greater_than_mean}")

# 4. Normalize the dataset using min-max scaling
normalized_data = (data - min_value) / (max_value - min_value)
print(f"First 5 Normalized Values: {normalized_data[:5]}")
```
