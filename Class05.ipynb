```python
stack = []

while True:
    print("\nChoose an operation:")
    print("1. Push an element onto the stack")
    print("2. Pop an element from the stack")
    print("3. Display the stack")

    try:
        choice = int(input("Enter your choice (1-3): "))
    except ValueError:
        print("Invalid input! Please enter a number between 1 and 3.")
        continue

    if choice == 1:
        element = input("Enter the element to push onto the stack: ")
        stack.append(element)
        print(f"Element {element} pushed onto the stack.")
    elif choice == 2:
        if not stack:
            print("Error: Cannot pop from an empty stack!")
        else:
            removed_element = stack.pop()
            print(f"Element {removed_element} popped from the stack.")
    elif choice == 3:
        if not stack:
            print("The stack is currently empty.")
        else:
            print("Current stack:", stack)
        break
    else:
        print("Invalid choice! Please enter a number between 1 and 3.")
```

```python
import random

numbers = [random.randint(1, 100) for i in range(10)]

maximum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num

minimum = numbers[0]
for num in numbers:
    if num < minimum:
        minimum = num

total = 0
for i in numbers:
    total += i
average = total / 10

print("Generated list of numbers:", numbers)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
print("Average value:", average)
```
